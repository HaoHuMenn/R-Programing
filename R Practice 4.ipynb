### Exercise 2
Create a list of formulas for modeling income with:
- *height
- *height * weight
- *linear combination of all variables

````r
# exercise 2
# income vs height
ggplot(heights, aes(height, income)) + geom_point()
# as we can see from the plot, height and income 
# is more of a categorical ralationship
````

````r

# formulas with categorical variables
model_matrix(heights, income ~ height)
````

````r
# quadratic model with imcome and height
model_matrix(heights, income ~ I(height^2))
````

````r
# logarithmic model
model_matrix(heights, income ~ log(height))
````

````r
# plot income with height*weight
ggplot(heights, aes(income, height*weight)) + geom_point()
# this one's tendacy looks like catigorical on different incomes 
````

````r
# liner formulas list:
model_matrix(heights, I(height*weight) ~ income)
````

````r
# linear combination of all variables, height, weight
model_matrix(heights, height ~ weight)
````

### Exercise 3
From the data, remove rows containing NA’s. Fit the linear model with the formulas from exercise 2.

````r
# exercise 3
new <- na.omit(heights)
# fit model
mod1 <- lm(income~height, data = new)
mod2 <- lm(height * weight ~ income, data = new)
````

### Exercise 4
For each fit, calculate ```RMSE```.

````r
# exercise 4
# find unique value of x
grid1  <- new %>% data_grid(height)
grid2 <- new %>% data_grid(income)
````

````r
# make predictions and add them to grid data frame
grid1 <- grid1 %>% add_predictions(mod1)
grid1
grid2 <- grid2 %>% add_predictions(mod2)
grid2
````

````r
# calculate RMSE
library(Metrics)
# RMSE for income predicted by height
rmse(new$income, predict(mod1, data.frame(height = new$height)))
````

````r
# RMSE for height*weight predicted by income
new$hw <- new$height*new$weight
rmse(new$hw, predict(mod2, data.frame(income = new$income)))
````
